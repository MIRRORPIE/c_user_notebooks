```python
<pro_02--------------------------------CODE--------------------------------->

package pro_02;
import java.util.Scanner;
public class Pro_02 {
    static boolean isPrime(int n) 
    { 
        // Corner cases 
        if (n <= 1) return false; 
        if (n <= 3) return true; 
        // This is checked so that we can skip middle five numbers in below loop 
        if (n % 2 == 0 || n % 3 == 0) return false;
        for (int i = 5; i * i <= n; i = i + 6) 
            if (n % i == 0 || n % (i + 2) == 0) 
            return false; 
        return true; 
    }
    
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        System.out.print("Enter any number: ");
        int n = scanner.nextInt();
        String msg = isPrime(n) ? "Prime." : "Not Prime.";
        System.out.println("Number " + n + " is " + msg);
    }  
}
<----------------------------------OUTPUT----------------------------------->
Enter any number: 17
Number 17 is Prime.
```

```python
Rectangle.java--------------------------class------------------------------->

package pro_04;
public class Rectangle {
    
    public float length;                     //Instance variable
    public float breadth;    
    
    public Rectangle() {                               // Default constructor
        System.out.println("Default Constructor Triggered.");
    }
    
    public Rectangle(float l, float b) {          // Parameterized constructor
        System.out.println("Parameterized Constructor Triggered.");
        this.length = l;                     // if we will not do this then 
        this.breadth = b;                    // r.length will show 0.0 in main.
        //this keyword refers to the current instance of the class
    }
}

<pro_04--------------------------------CODE--------------------------------->

package pro_04;
public class Pro_04 {
    public static void main(String[] args) {
        
        Rectangle r1 = new Rectangle();    
        /* it creates new object inside the heap memory. and it 
           calls the default construcor.
        */                         
        r1.breadth = 8.7f;
        r1.length = 12.9f;
        System.out.printf("%.2f\t%.2f\n", r1.length, r1.breadth);
        
        Rectangle r2 = new Rectangle(14.6f, 20.8f);
        // it triggers the parameterized constructor.
        System.out.printf("%.2f\t%.2f\n", r2.length, r2.breadth);
    } 
}
<----------------------------------OUTPUT----------------------------------->

Default Constructor Triggered.
12.90   8.70
Parameterized Constructor Triggered.
14.60   20.80
```

```python
Rectangle.java--------------------------class------------------------------->

package pro_04;
public class Rectangle {
    
    public float length;      
    public float breadth;     
    
    public Rectangle() {      
        this( 10.5f, 20.4f);
        System.out.println("Default Constructor Triggered.");
        // this( 10.5f, 20.4f);       // here we cannot put...should be in 1st line
    }
    
    public Rectangle(float length) {
        System.out.println("Constructor 02 Triggered");
        this.length = length;
    }
    
    public Rectangle(float length, float breadth) {      
        System.out.println("Parameterized Constructor Triggered.");
        this.length = length;
        this.breadth = breadth;    
    }
}

<pro_04--------------------------------CODE--------------------------------->

package pro_04;
public class Pro_04 {
    public static void main(String[] args) {
        
        Rectangle r1 = new Rectangle();    
        System.out.printf("%.2f\t%.2f\n", r1.length, r1.breadth);
        
        Rectangle r2 = new Rectangle(14.6f, 20.8f);
        // it triggers the parameterized constructor.
        System.out.printf("%.2f\t%.2f\n", r2.length, r2.breadth);
    } 
}
<----------------------------------OUTPUT----------------------------------->

Parameterized Constructor Triggered.
Default Constructor Triggered.
10.50	20.40
Parameterized Constructor Triggered.
14.60	20.80
```

### Program_08A:=======abstract keyword================>

```python
<pro_08--------------------------------CODE--------------------------------->

// abstract keyword:------------------------------>

package pro_08;
public class Pro_08 {
    public static void main(String[] args) {
        //Animal animal = new Animal();   // Not allowed
        Animal animal;     // but we can create reference of abstract class.
        Animal animal1 = new Dog();  // Parent class reference --> child class object
        animal1.run();
        animal1.eat();
    }
}
abstract class Animal {
    public void run() {
        System.out.println("Animal is running.");
    }
    abstract public void eat();   // No method body
}
class Dog extends Animal{
    @Override
    public void eat() {
        System.out.println("Dog is eating.");
    }
}

/*
    Rules for abstract keyword:
        1. abstract class:
            * A class that is declared abstract
            * u cannot create object of abstract class
            * It may or may not contain abstract methods.

        2. abstract method:
            * No method body. you cannot write code in abstract method.
            * It is mandatory to override the abstract method in child class.
*/
<----------------------------------OUTPUT----------------------------------->
Animal is running.
Dog is eating.
```

### Program_9:============Interface=================>

```python
<pro_09A--------------------------------CODE--------------------------------->

// Interface :------------------>

package pro_09;
public class Pro_09 {
    public static void main(String[] args) {
        // Animal animal = new Animal();    // not valid.
        Animal animal1;
        Animal animal2 = new Dog();
        animal2.eat();
        animal2.run();
    }
}

interface Animal {
    public static final float PI = 3.14f;    // final ---> we cannot change the 
                                              // value of pi.not good to use variables
    abstract public void run();
    abstract public void eat();
}

class Dog implements Animal {
    
    int x = 10;
        
    @Override
    public void run() {
        System.out.println("Dog is running.");
    }
    
    @Override
    public void eat() {
        System.out.println("Dog is eating.");
    }
}

/*
Rules of Interface:---->
    * U cannot create object of interface. They are just a blueprint of a class.
    * Use implement keyword
    * There can only contain abstract methods.

    1. For Methods:->
        * They are public and abstract
        * No method body. U cannot write code within ur methods.

    2. For Variables:->
        * Avoid using field variables.
        * But when used they r constants. Their values cannot be changed once defined.
        * They r public, static and final.
*/
<----------------------------------OUTPUT----------------------------------->
Dog is eating.
Dog is running.
```

```python
<pro_09B--------------------------------CODE--------------------------------->
// Interface:--->

package pro_09;
public class Pro_09 {
    public static void main(String[] args) {
        Remote remote = new Television();
        remote.powerOn();
        remote.volumeUp();
        remote.powerOff();
    }
}

interface Remote {
    abstract public void volumeUp();     // we can remove abstract public..
    abstract public void volumeDown();   // because in interface, 
    abstract public void powerOff();     // There can only contain abstract methods.
    abstract public void powerOn();
}

class Television implements Remote{
    @Override
    public void volumeUp() {
        System.out.println("Volume Up");
    }
    @Override
    public void volumeDown() {
        System.out.println("Volume Down");
    }
    @Override
    public void powerOff() {
        System.out.println("Power Off");
    }
    @Override
    public void powerOn() {
        System.out.println("Power On");
    } 
}
<----------------------------------OUTPUT----------------------------------->
Power On
Volume Up
Power Off
```

* Abstract Class
    * The abstract keyword is used to declare an abstract class.
    * We use extends keyword for inheritance.
    * An Abstract class can have abstract and non-abstract methods.
    * Abstract class doesn't support multiple inheritance.
    * An Abstract class can have final, non-final, static and non-static variables.

* Interface
    * The interface keyword is used to declare an interface.
    * We use implements keyword for implementing an interface in a class.
    * An Interface can have only abstract methods.
    * An Interface supports multiple inheritance.
    * An Interface has only static and final variables.

### Program_12:========File Handling=================>

```python
<pro_12--------------------------------CODE--------------------------------->
// File operation:----------->write and read data:---->

package pro_12;
import java.io.FileInputStream;
import java.io.FileOutputStream;
import java.io.IOException;

public class Pro_12 {
    public static void main(String[] args) {
        
        String data = "I Love India.";
        // Write data
        try {
            FileOutputStream fos = new FileOutputStream("myfile1.txt");
            byte[] byteData = data.getBytes();
            fos.write(byteData);
            fos.close();
        } catch (IOException ex) {
            System.out.println(ex);
        }
        
        // Read data
        try {
            FileInputStream fis = new FileInputStream("myfile1.txt");
            int i = 0;
            while ( ( i = fis.read() ) != -1) {
                System.out.print((char) i);
            }
            fis.close();
        } catch (IOException ex) {
            System.out.println(ex);
        }
    }
}
<----------------------------------OUTPUT----------------------------------->
I Love India.
```

```python
//NAME:--------------------------------><app_layout.fxml>
    
<?xml version="1.0" encoding="UTF-8"?>

<?import javafx.scene.control.*?>
<?import javafx.scene.layout.*?>
<?import javafx.scene.text.*?>

<VBox maxHeight="-Infinity" maxWidth="-Infinity" minHeight="-Infinity" minWidth="-Infinity" prefHeight="266.0" prefWidth="258.0" xmlns="http://javafx.com/javafx/10.0.2-internal" xmlns:fx="http://javafx.com/fxml/1" fx:controller="App_02_FX_temp.Controller">
    <children>
        <Pane prefHeight="273.0" prefWidth="258.0">
            <children>
                <Label fx:id="welcomeLabel" alignment="CENTER" layoutX="1.0" layoutY="29.0" prefHeight="27.0" prefWidth="258.0" text="Welcome">
                    <font>
                        <Font name="System Bold" size="18.0" />
                    </font>
                </Label>
                <ChoiceBox fx:id="choiceBox" layoutX="41.0" layoutY="70.0" prefHeight="25.0" prefWidth="178.0" />
                <TextField fx:id="userInputField" layoutX="41.0" layoutY="115.0" prefHeight="25.0" prefWidth="178.0" promptText="Enter Temperature" />
                <Button fx:id="convertButton" layoutX="80.0" layoutY="165.0" mnemonicParsing="false" prefHeight="25.0" prefWidth="100.0" text="Convert" />
            </children>
        </Pane>
    </children>
</VBox>
```

```python
//NAME:--------------------------------><Controller.java>
    
package App_02_FX_temp;

import javafx.beans.value.ChangeListener;
import javafx.beans.value.ObservableValue;
import javafx.event.ActionEvent;
import javafx.event.EventHandler;
import javafx.fxml.FXML;
import javafx.fxml.Initializable;
import javafx.scene.control.*;

import java.net.URL;
import java.util.ResourceBundle;

public class Controller implements Initializable {

	@FXML
	public Label welcomeLabel;

	@FXML
	public ChoiceBox<String> choiceBox;

	@FXML
	public TextField userInputField;

	@FXML
	public Button convertButton;

	private static final String C_TO_F_TEXT = "Celsius to Fahrenheit";
	private static final String F_TO_C_TEXT = "Fahrenheit to Celsius";

	private boolean isC_TO_F = true;

	@Override
	public void initialize(URL location, ResourceBundle resources) {

		choiceBox.getItems().add(C_TO_F_TEXT);
		choiceBox.getItems().add(F_TO_C_TEXT);

		choiceBox.setValue(C_TO_F_TEXT);

		choiceBox.getSelectionModel().selectedItemProperty().addListener((observable, oldValue, newValue) -> {
			System.out.println(newValue);

			if(newValue.equals(C_TO_F_TEXT)) {
				isC_TO_F = true;
			}
			else{
				isC_TO_F = false;
			}

		});

		convertButton.setOnAction(event -> {
			System.out.println("Button Clicked");
			
			convert();
		});
	}

	private void convert() {

		String input = userInputField.getText();    // 23.4 ===> "23.4"

		float enteredTemperature = 0.0f;

		try{
			enteredTemperature = Float.parseFloat(input);
		} catch (Exception ex) {
			warnUser();
			return;
			// no code executed
		}

		Float newTemperature = 0.0f;

		if(isC_TO_F) {
			newTemperature = (enteredTemperature * 9 / 5) + 32;
		} else {
			newTemperature = (enteredTemperature - 32) * 5 / 9;
		}

		display(newTemperature);
	}

	private void warnUser() {
		Alert alert = new Alert(Alert.AlertType.ERROR);
		alert.setTitle("Error Occured");
		alert.setContentText("Please enter a valid temperature");
		alert.show();
	}

	private void display(float newTemperature) {

		String unit = isC_TO_F ? " F" : " C";
		System.out.println("The new Temperature is: " + newTemperature + unit);

		Alert alert = new Alert(Alert.AlertType.INFORMATION);
		alert.setTitle("Result");
		alert.setContentText("The new temperature is: " + newTemperature + unit);
		alert.show();

	}
}
```

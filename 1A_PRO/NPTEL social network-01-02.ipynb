* Degree distribution:-
    no. of edges to the node

```python
#find out the degree distribution

import networkx as nx
import matplotlib.pyplot as plt

def plot_deg_dist(G):
    all_degrees=list(dict(nx.degree(G)).values())
    unique_degrees=list(set(all_degrees))
    
    count_of_degrees=[]
    for i in unique_degrees:
        x=all_degrees.count(i)
        count_of_degrees.append(x)
        
    plt.plot(unique_degrees,count_of_degrees,"ro-")
    plt.xlabel("Degrees--------------->")
    plt.ylabel("number of nodes------------->")
    plt.title("<-degree distribution of karate network->")
    plt.axis([0,20,0,15])
    plt.show()

G=nx.read_gml("html_folder/karate.gml",label="id")

print(nx.info(G),"\n")
print(nx.degree(G),"\n")

plot_deg_dist(G)


#nx.draw(G,with_labels=1)
#plt.show()
```

```python

import networkx as nx
import random

def add_nodes(n):    #add n no. of nodes to the graph
    G=nx.Graph()
    G.add_nodes_from(range(n))
    return G

def add_random_edge(G):   #add random edges
    v1=random.choice(list(G.nodes()))
    v2=random.choice(list(G.nodes()))
    #print(v1,v2)
    if(v1!=v2):
        G.add_edge(v1,v2)
    return G

def add_till_connectivity(G):
    while(nx.is_connected(G)==False):
        G=add_random_edge(G)
    return(G)

n=int(input("enter the number of nodes:"))
G=add_nodes(n)
G=add_till_connectivity(G)
print(G.edges(),"\n")
print("G.number_of_nodes()=",G.number_of_nodes())
print("G.number_of_edges()=",G.number_of_edges(),"\n")
print(nx.info(G))
#nx.draw(G,with_labels=1)
```

```python
# programming illustration of above proof:->

import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy

def add_nodes(n):    #add n no. of nodes to the graph
    G=nx.Graph()
    G.add_nodes_from(range(n))
    return G

def add_random_edge(G):   #add random edges
    v1=random.choice(list(G.nodes()))
    v2=random.choice(list(G.nodes()))
    #print(v1,v2)
    if(v1!=v2):
        G.add_edge(v1,v2)
    return G

def add_till_connectivity(G):
    while(nx.is_connected(G)==False):
        G=add_random_edge(G)
    return(G)

def create_instance(n):
    G=add_nodes(n)
    G=add_till_connectivity(G)
    return(G.number_of_edges())

def create_avg_instance(n):   #average it over 50 instances
    list1=[]
    for i in range(100):
        list1.append(create_instance(n))
    return(numpy.average(list1))

def plot_connectivity():
    x=[]
    y=[]
    i=10
    while(i<=200):
        x.append(i)
        y.append(create_avg_instance(i))
        i+=10
    plt.xlabel("Number of nodes:----->")
    plt.ylabel("No. of edges required to connect the graph:---->")
    plt.title("<-:Emergence of connectivity:->")
    plt.plot(x,y,"ro-")
    
    x1=[]
    y1=[]
    i=10
    while(i<=200):
        x1.append(i)
        y1.append(i*numpy.log(i))   #doing the nlogn experiment
        i+=10
    plt.xlabel("Number of nodes:----->")
    plt.ylabel("No. of edges required to connect the graph:---->")
    plt.title("<-:Emergence of connectivity:->")
    plt.plot(x1,y1,"yo-")
    
    y2=[i/2 for i in y1]
    plt.plot(x1,y2,"bo-")
    plt.show()
        

plot_connectivity()
```
